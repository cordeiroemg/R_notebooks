* When inputs (i.e., dependent variable, predictors or features) are used to predict the values of the outputs (i.e., dependent variable or responses), we call it supervised learning. 

## Variable Types and Terminology

This distinction in input type (X):
* **quantitative measurement**; some measures are bigger than others
* **qualitative measurements**; assumes values in a finite set. Also refered as **categorical** or **discrete** variables as well as **factors**
    . Often represented as codes or two classes. Also refered as targets.
    . Can be codified as dummy variables
* We can have bot quantitative and qualitative measurements used as predictors
* **ordered categorical**; when there is ordering among such as small, medium, and large.

This distinction in output type (Y or G):
* **regression** when we predict quantitative outputs
* **classification** when we predict qualitative outputs

## Two Simple Approaches to Prediction: Least Squares and Nearest Neighbors

Two of the most important linear models:
  * (1) Least squares (great assuptions but overall stable)
  * (2) k-nearest-neighbor (little assuptions but can be unstable)
  
 



After finding $ \hat β $, then we can formulate that an arbitrary $x_0$ will predict a $ \hat y $ as:

<p>&nbsp;</p>
\begin{split}
\hat y(x_0) = x_i^T \hat β
\end{split}
<p>&nbsp;</p>

and the estimated covariances for linear regression coefficient estimates are:

<p>&nbsp;</p>
\begin{split}
\hat {Var}(\hat β) = \hat σ^2 (X^T X)^{-1}
\end{split}
<p>&nbsp;</p>


* Classification problems have two distinct scenarios:

 * (1) Scenario 1: training data in each class were generated from bivariate Gaussian distributions with uncorrelated components and different means

 * (2) Scenario 2: training data in each class came from a mixture of 10 low-variance Gaussian distribution, with individual means themselves distributed as Gaussian.

In the case of mixtures of tightly clustered Gaussians the story is different. A linear decision boundary is unlikely to be optimal, and in fact is not. The optimal decision boundary is nonlinear and disjoint, and as such will be much more difficult to obtain.

## Generating simulated data

* (1) generated 10 means $m_k$ from a bivariate Gaussian distribution $N((1, 0)^T, I)$ for blue and $N((0, 1)^T, I)$ for the yellow data

* Then for each class we generated 100 observations as follows: for each observation, we picked an $m_k$ at random with probability 1/10, and then generated a $N(m_k , I/5)$

```r
set.seed(1)
data.sim = function(N, b.xy, o.xy){
        # Generating mk sets size 10 for each color group
        blue.mk = as.data.frame(mvrnorm(n = 10, mu = b.xy, Sigma= diag(2)))
            blue.mk$labels = factor(x = rep(0,nrow(blue.mk)), levels = c(0,1), labels = c("blue", "orange"))


        orange.mk = as.data.frame(mvrnorm(n = 10, mu = o.xy, Sigma= diag(2)))
            orange.mk$labels = factor(x = rep(1,nrow(orange.mk)), levels = c(0,1), labels = c("blue", "orange"))

        # sample from mk and generate a distribution for blue and orange of size N

        blue.sample.mk = blue.mk[sample(1:nrow(blue.mk), N, replace = TRUE), 1:2]
        orange.sample.mk = orange.mk[sample(1:nrow(orange.mk), N, replace = TRUE), 1:2]

        blue.data = as.data.frame(t(apply(blue.sample.mk, MARGIN = 1, function(x) mvrnorm(n = 1, mu = x, Sigma = diag(2)/5))))
            blue.data$labels = factor(x = rep(0, nrow(blue.data)), levels= c(0,1), labels = c("blue", "orange"))

        orange.data = as.data.frame(t(apply(orange.sample.mk, MARGIN = 1, function(x) mvrnorm(n = 1, mu = x, Sigma = diag(2)/5))))
            orange.data$labels = factor(x = rep(1, nrow(orange.data)), levels= c(0,1), labels = c("blue", "orange"))

        # Generating final table
        training.df = rbind.data.frame(blue.data, orange.data)
        training.df = cbind.data.frame(1, training.df)                                    
        colnames(training.df)= c("intercept", "x", "y", "labels")                               
        return(training.df)
}
```

## Linear Model and Least Squares


Given an vector inputs $ X^T = (X_1, X_2, ..., X_p)$, we can predict the output *Y* via the model:

<p>&nbsp;</p>
\begin{split}
\hat Y & = \hat β_0 + \sum_{j=1}^p X_j \hat β_0
\end{split}
<p>&nbsp;</p>

The term $ \hat β_0 $ is the intercept, also called "bias" in machine learning

<p>&nbsp;</p>
\begin{split}
\hat Y & = X^T \hat β
\end{split}
<p>&nbsp;</p>
  
 If we are modeling a single output, $ \hat Y $ is a scalar. In general, $ \hat Y $ can be a K-vector, in which $β$ would be a $ p * K$ matrix of coefficients. In the (p + 1)-dimensions input-output space, $(X, \hat Y)$ represents a hyperplane.

### How to fit a linear model?

* There are different methods, but the most popular is the **method of least squares**. The coefficients $β$ is calculated in a way to minimize the residual sum of squares (RSS)


<p>&nbsp;</p>
\begin{split}
RSS(β)  = \sum_{i=1}^{N}(y_i - x_i^Tβ)^2
\end{split}
<p>&nbsp;</p>
  
$RSS(β)$ is a quadratic function, thus always has the minimum point, but it might not be unique. It is more easely to calculate in its matricial form:

<p>&nbsp;</p>
\begin{split}
RSS(β)  = (y - xβ)^T (y - xβ)
\end{split}
<p>&nbsp;</p>

Where $X$ is an $N * p$ matrix with each row an input vector, and y is an N-vector of the outputs in the training set. 

If we differentiate with restect to $β$
<p>&nbsp;</p>
\begin{split}
X^T(y - Xβ)  = 0
\end{split}
<p>&nbsp;</p>

If $ X^TX $ is a nonsingular matrix,  then the values of $\hat β$ that will minize squared error between y and tiffet $X*β$ are:
<p>&nbsp;</p>
\begin{split}
\hat β = (X^T X)^{-1}X^Ty 
\end{split}
<p>&nbsp;</p>

We can estimate the variance of the residual noise $ \hat σ^2 $ (mean square error as:

<p>&nbsp;</p>
\begin{split}
\ σ^2 = \frac{(y - X \hat β)^T(y - X \hat β)}{n - p - 1} 
\end{split}
<p>&nbsp;</p>



### (1) Finding beta

\begin{split}
\hat β = (X^T X)^{-1}X^Ty 
\end{split}
<p>&nbsp;</p>

```r
manual.lm = function(X, y){
    return(beta.hat = solve(t(X) %*% X) %*% t(X) %*% y)
}
```

```r
X = as.matrix(train[,1:2])
y = as.matrix(train[,3])

beta.hat = manual.lm(X, y)
beta.hat
```

### (2) Finding the residual 

\begin{split}
y - X \hat β 
\end{split}
<p>&nbsp;</p>

```r
residuals <- y - (X %*% beta.hat)
df.residuals <- data.frame(residuals)


ggplot(data = df.residuals) +
    geom_histogram(mapping = aes(residuals), binwidth = 0.3, fill = "#D95F02") +
    setting 
```

### (3) Finding the (residual) variance-covariance matrix

<p>&nbsp;</p>
\begin{split}
\hat {Var}(\hat β) = \hat σ^2 (X^T X)^{-1}
\end{split}
<p>&nbsp;</p>

```r
# setting degree of freedom
p <- 2 - 1
n <- nrow(X)

residual.var <- t(residuals) %*% residuals / (n - p - 1)
residual.var <- as.numeric(residual.var)
#residual.var 

# Variance-covariance matrix
beta.covar <- residual.var * X
head(beta.covar)

# diag takes the diagonal of the matrix, sqrt makes it go
# from variance to standard deviation
beta.SE <- sqrt(diag(beta.covar))
round(beta.SE,3)
```

## Nearest-Neighbor Methods 

Nearest-neighbor methods use those observations in the training set $T$ closest in input space to $x$ to form $ \hat Y$ .

<p>&nbsp;</p>
\begin{split}
\hat Y(x) = \frac{1}{k} \sum_{x_i \epsilon N_k(x)} y_i
\end{split}
<p>&nbsp;</p>

* In summary, we find the $k$ observations with $x_i$ closest to $x$ in input space, and average their responses

The algorith is as follows:
   * 1) Determine a value for parameter k
   * 2) Calculate the distance between the query-instance of all training samples
   * 3) Sorting distance and determine the nearest neightbors base of the k-th minimum distance
   * 4) Gather the category of the nearest neighbors
   * 5) Use simple majority of the category of nearest neighbors as the prediction value of the query instance
