# Linear regression: the basics

* Summarizes how the average values of a numerical *outcome* variable vary over subpopulations defined by linear function of *predictors*
* It can be thought as a way to compare across predicted values or as comparisons among averages in the data rather than just explore the relation between variables

## One predictor

### Binary predictor
* For a binary predictor, the regression coefficient is the difference between the averange of the two groups (i.e., mother finished highschool '1', or did not finish highschool '0')

```r
library(foreign)
library(ggplot2)
library(cowplot)
library(ggsci)
library(dplyr)

setwd("/home/erickcordeiro/Data_Analysis_Using_Regression_and_Multilevel_Hierarchical_Models/ARM_Data")
setting <- theme_linedraw(base_size = 18)
```

```r
kidiq = read.dta(file="./child.iq/kidiq.dta")
kidiq$mom_hs = as.factor(kidiq$mom_hs)
head(kidiq)
```

```r
ggplot(data = kidiq, mapping = aes(x= mom_hs, y = kid_score)) +
    geom_jitter(width=0.25, size = 3) + xlab("Mom finished high school") + ylab("Kid's scores") + setting
```

```r
fit.1 = lm(kid_score ~ mom_hs)
summary(fit.1)
```

<p>&nbsp;</p>
\begin{split}
\hat {Kid.score} = 78 + 12 mom.hs + error
\end{split}
<p>&nbsp;</p>

* The slope (12) is the average score between the two groups (90 - 78 = 12)
* The intercept (78) is the average score for whose mothers did not complete high school

```r
ggplot(data = kidiq, mapping = aes(x= mom_hs, y = kid_score)) +
    geom_abline(intercept = 78, slope = 12) +
    geom_jitter(width=0.25, size  = 3) + xlab("Mom finished high school") + ylab("Kid's scores") + setting 
```

### Continous predictor

* The points in the line are the predicted test scores for children at the several maternal IQ levels or the average test scores for subpopulations 
* If we increase the mother IQ groups by 1 point, the average score of the kid increased 0.6 points or for every  10 points increase in the mom's IQ, kid's have a 6 points increment in their scores
* The intercept refers to the kids whose mom's have IQ zero (not observable)

```r
ggplot(kidiq, mapping = aes(x = mom_iq, y = kid_score)) +
        geom_point(size=3) +
        xlab("Mom'IQ score") + ylab("Kid's scores") + setting
```

```r
fit.2 = lm(kid_score ~ mom_iq)
summary(fit.2)
```

<p>&nbsp;</p>
\begin{split}
\hat {Kid.score} = 26 + 0.6 mom.iq + error
\end{split}
<p>&nbsp;</p>

* The slope (0.6) is the rate of increase in score for every unit increase in mom's IQ
* The intercept (26) is the average score for kids whose mothers scored '0' in their IQ test


```r
ggplot(kidiq, mapping = aes(x = mom_iq, y = kid_score)) +
        geom_point(size=3) +
        geom_abline(intercept = 26, slope = 0.6) +
        xlab("Mom'IQ score") + ylab("Kid's scores") + setting
```

## Multiple predictors

* Regression coefficients are more complicate to interpret with multiple predictors due to the interelation between variables
* It is often helpful held all other coefficients constant while interpreting a given predictor

```r
fit.3 = lm(kid_score ~ mom_hs + mom_iq)
summary(fit.3)
```

<p>&nbsp;</p>
\begin{split}
\hat {Kid.score} = 26 + 6 mom.hs + 0.6 mom.iq + error
\end{split}
<p>&nbsp;</p>

*  This model forces the slope of the regression of child’s test score on mother’s IQ score to be the same for each maternal education subgroup (high school complete and high school incomplete).

```r
ggplot(kidiq, aes(aes(x = mom_iq, y = kid_score, group = mom_hs))) +
        geom_point(aes(x = mom_iq, y = kid_score, color = mom_hs), size = 3, show.legend = FALSE) +
        geom_abline(intercept = 26, slope = 0.6, color = "black", size = 1.5) +
        geom_abline(intercept = (26+12), slope = 0.6, color = "darkgray", size = 1.5) +
        xlab("Mom'IQ score") + ylab("Kid's scores") + setting +  scale_color_jco()
```

* **Intercept**: if a child had a mother with an IQ of zero that did not complete the high school, then the child's average score is 26
* **coefficient of maternal high school completion**: the model predicts an expected difference of 6 points between kids whose mother's completed high school and those who their mother did not complete high school

### Counterfactual and predictive interpretations

Distinstion on interpreting the regression coefficients:

**predictive interpretations**: the coefficient is expected to differ in $y$ between the two groups


**counterfactual interpretation**: changes within individuals. Ex: increase from 100 to 101 in mom's IQ, increases child's childs score in 0.6

## Interactions

* The slope of the regression of child's test score on mother's IQ was forced to be equal across subgroups (high school completion) but they can be substantially different
* The solution for that is to include an interaction between $mom.hs$ and $mom.iq$

```r
fit.4 = lm(kid_score ~ mom_hs * mom_iq)
summary(fit.4)
```

<p>&nbsp;</p>
\begin{split}
\hat {Kid.score} = -11 + 51 mom.hs + 1.1 mom.iq - 0.5   interaction  + error
\end{split}
<p>&nbsp;</p>


```r
ggplot(kidiq, aes(aes(x = mom_iq, y = kid_score, group = mom_hs))) +
        geom_point(aes(x = mom_iq, y = kid_score, color = mom_hs), size = 3, show.legend = FALSE) +
        geom_abline(intercept = -11, slope = 1.1, color = "black", size = 1.5) +
        geom_abline(intercept = (51-11), slope = 0.6, color = "darkgray", size = 1.5) +
        xlab("Mom'IQ score") + ylab("Kid's scores") + setting +  scale_color_jco()
```

**Intercept**: prediction test score for children whose mothers did not complite high school and had IQs of 0
**Coefficient of high school completion**: expected score difference between two groups
**coefficient of mom's IQ**: mean test scores across children whose mothers did not complete high school, but whose mother differ by 1 point in IQ
**Coefficient of the interaction**:represent the *difference* in the slope for mom.iq

### When should we look for interactions?

* Interaction can be more easily interpreted if during the pre-process we center each input variable about its mean (or some other convenient reference point)
