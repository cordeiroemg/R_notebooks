## What is multilevel regression modeling?


* A model in which the coefficients vary by school, $y = \alpha_j + \beta_jx + error$ where the subscripts $j$ index schools
* A model with more than one variance component (student-level and school-level variation)
* Regression with many predictors, including an indicator variable for each school in the data

Multilevel model: regression in which the paramters (regression coefficients) are given a probability model

Two key parts of a multilevel  model are varying coefficients and a model for those varying coefficients

### Models for regression coefficients

Assume two multilevel models; one **student-level** predictor $x$ (pre-test score) and one **school-level** predictor $u$ (average parent's incomes)

#### Varying-intercept model
Those regressions have the **same slope** in each school and only the **intercepts vary**

For students $i=1, ..., n$
<p>&nbsp;</p>
\begin{split}
y_i = \alpha_j[i]+\beta x_i + \epsilon_i 
\end{split}
<p>&nbsp;</p>

For schools $j=1,...,J$
<p>&nbsp;</p>
\begin{split}
\alpha_j = \alpha + b u_j + \eta_j 
\end{split}
<p>&nbsp;</p>

where $x_i$ and $u_j$ are the predictors at the student and school levels. And $\epsilon_i$ and $\eta_j$ are the independent error terms at each of the two levels

#### Varying-intercept, varying-slope model

More complex. Both intercept and slope vary

For students $i=1, ..., n$
<p>&nbsp;</p>
\begin{split}
y_i = \alpha_j[i]+\beta_j[i] x_i + \epsilon_i 
\end{split}
<p>&nbsp;</p>

For schools $j=1,...,J$
<p>&nbsp;</p>
\begin{split}
\alpha_j = \alpha_0 + b_0 u_j + \eta_j1 
\end{split}
<p>&nbsp;</p>

and 
<p>&nbsp;</p>
\begin{split}
\beta_j = \alpha_1 + b_1 u_j + \eta_j2 
\end{split}
<p>&nbsp;</p>

#### Labels 
Multilevels or hierarchical. Multi level with parameters and hyperparameters. 

* Cases of non-nested models, individual observations that are nested within states and years. Neither "states" nor "year" is above the other in hierarchical sense

Multilevel models are often nown as random-effects or mixed-effects models. The regression coefficients that are being modeled are called *random effects*. *Fixed effects* correspond either to parameters that do not vary and are being tested. *Mixed-effects* are models that include both fixed and random effects

#### Some examples from our own research

* Multiple measures within the house
* Houses
* Counties

For house measures $i=1, ..., n$
<p>&nbsp;</p>
\begin{split}
log(y_i) = \alpha_j[i]+\beta x_i + \epsilon_i \: \text{within county variation}
\end{split}
<p>&nbsp;</p>

$x_i$ is the floor of the measurement (0, 1)

For schools $j=1,...,J$
<p>&nbsp;</p>
\begin{split}
\alpha_j = \alpha + b u_j + \eta_j \:  \text{between county variation}
\end{split}
<p>&nbsp;</p>

$u$ is the uranium measurement at the county level
