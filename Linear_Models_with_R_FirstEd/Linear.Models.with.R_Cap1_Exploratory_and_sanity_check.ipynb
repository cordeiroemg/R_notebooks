#### 768 adult female Pima Indians living near Phoenix. Variables recorded: 
* number of times pregnant
* plasma glucose concentration at 2 hours in an oral glucose tolerance test
* diastolic blood pressure (mmHg)
* triceps skin fold thickness (mm), 2-hour serum insulin (mu U/ml)
* body mass index (weight in kg/(height in m 2 ))

* age (years)

### When to use Regression Analysis

Used to explain or model the relationship between a single variable $Y$, called response, output or dependent variable; and one called predictor, input, independent , or explanatory variable $X_1,..., X_p$ when $p = 1$ it is called **simple regression**. When $p>1$, then it is called **multiple regression** or **multivariate regression**. When there is more than one $Y$, we called **multivariate multiple regression**.

* The response variable **must** be continuous
* The explanatory variable can be continous, discrete or categorical

If X is only quantitative -> regression
If X is both quantitative and qualitative -> analysis of covariance
If X is only qualitative -> analysis of variance (ANOVA)

If the reponse variable is qualitative -> logistic regression could be used

### Objectives of a regression analysis 

1) Predic the future observation
2) Assessment of the effect of, or relationship between variables and the response
3) A general description of data structure

```r
data(stat500)
stat500 = data.frame(scale(stat500))
plot(final ~ midterm, stat500)
# we add the y=x for a parameter of perfect association
abline(0,1)
g <- lm(final ~ midterm, stat500)
abline(coef(g), lty = 5)
cor(stat500)
```
