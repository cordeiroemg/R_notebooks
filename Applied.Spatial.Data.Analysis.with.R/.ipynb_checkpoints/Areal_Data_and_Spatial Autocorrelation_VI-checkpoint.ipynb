# Areal Data and Spatial Autocorrelation


## Introduction

* Continuous  data  involves  point samples from a continuous spatial  distribution  (i.e., temperature , Zn concentration, etc),  areal  data  involves aggregated  quantities  for  each  areal  unit  within  some  relevant  spatial  partition  of  a  given  region  (i.e.,  census  tracts  within  a  city,  or  counties  within  a  state). 

* Areal units may themselves constitute the units of observation.
* So positive spatial dependence tends to reduce the amount of information contained in the observations, because proximate observations can in part be used to predict each other.

* We have seen how distance on a continuous surface can be use to structure spatial autocorrelation with the variogram. Now we will be concerned with areal entities that are defined as neighbours.

* There are varies ways to define what a neighbour constitute and how to weight them. After that, we can measure spatial autocorrelation.
* asymmetry is present when i is a neighbour of j but j is not a neighbour of i

## Spatial Neighbours

* Creating spatial weights is a necessary step in using areal data, perhaps just to check that there is no remaining spatial patterning in residuals.

Steps:
   1. Define which relationships between observations are to be given a non-zero weight (choose the neighbour criterion)
   2. Assign weights to the identified neighbour links

#### Creating Graph-Based Neighbours

* The centroids, which is not the average of the coordinates in each dimension, are often take as point of reference to establish neighbours. 
* The most direct graph representation is the Delaunay triangulation of the points where the neighbour relations are defined by the triangulation.
* We can thin the triangulation as a Sphere of Influence (SOI) graph by removing links that are relatively long

```r
coords <- coordinates(Syracuse)
IDs <- row.names(as(Syracuse, "data.frame"))

library(tripack)

# Sphere of Influence (SOI)
Sy4_nb <- tri2nb(coords, row.names = IDs)
Sy5_nb <- graph2nb(soi.graph(Sy4_nb, coords), row.names = IDs)
plot(Syracuse, border = "grey60")
plot(Sy5_nb, coords, add = TRUE)
```

```r
# Gabriel graph neighbours
Sy6_nb <- graph2nb(gabrielneigh(coords), row.names = IDs)
plot(Syracuse, border = "grey60")
plot(Sy6_nb, coords, add = TRUE)
```

```r
# relative graph neighbours
Sy7_nb <- graph2nb(relativeneigh(coords), row.names = IDs)
plot(Syracuse, border = "grey60")
plot(Sy7_nb, coords, add = TRUE)
```

#### Distance-Based Neighbours


An alternative method is to choose the *k* nearest points as neighbours. This solution will adapt for different context in the region. It might, however, lead to a great number of asymmetric neigoubours, but it will ensure that all areas have *k* neighbours.

```r
Sy8_nb <- knn2nb(knearneigh(coords, k = 1), row.names = IDs)
Sy9_nb <- knn2nb(knearneigh(coords, k = 2), row.names = IDs)
Sy10_nb <- knn2nb(knearneigh(coords, k = 4), row.names = IDs)
nb_l <- list(k1 = Sy8_nb, k2 = Sy9_nb, k4 = Sy10_nb)


sapply(nb_l, function(x) is.symmetric.nb(x, verbose = FALSE, force = TRUE))
sapply(nb_l, function(x) n.comp.nb(x)$nc)
```

```r
par(mfrow = c(1, 1))

# K=1
plot(Syracuse, border = "grey60")
plot(Sy8_nb, coords, add = TRUE)


# K=2
plot(Syracuse, border = "grey60")
plot(Sy9_nb, coords, add = TRUE)

# K=4
plot(Syracuse, border = "grey60")
plot(Sy10_nb, coords, add = TRUE)
```

* We can use **nbdists** function to calculate a list of vectors of distance corresponding to the neighbour object (here for the first nearest neighbours).
* The greatest value will be the mininum distance needed to connect all locations with at least one neighbour.
* **dnearneigh** function is used to find neighbours with an interpoint distance. The arguments **d1** and **d2** set the lower and upper distance bounds. 
* Distance method is usually a problem when one of the nearest neighbour distance is much larger than other. In this case, distance must be set in a way guarantee that many areas have many neighbours.

```r
# calculate vector with neighbour distances 
dsts <- unlist(nbdists(Sy8_nb, coords))
summary(dsts)
```

```r
max_1nn <- max(dsts)
max_1nn
```

```r
# 75% of the maximum distance
Sy11_nb <- dnearneigh(coords, d1 = 0, d2 = 0.75 * max_1nn, row.names = IDs)
plot(Syracuse, border = "grey60")
plot(Sy11_nb, coords, add = TRUE)

# 100% of the maximum distance
Sy12_nb <- dnearneigh(coords, d1 = 0, d2 = 1 * max_1nn, row.names = IDs)
plot(Syracuse, border = "grey60")
plot(Sy12_nb, coords, add = TRUE)

# 150% of the maximum distance
Sy13_nb <- dnearneigh(coords, d1 = 0, d2 = 1.5 * max_1nn, row.names = IDs)
plot(Syracuse, border = "grey60")
plot(Sy13_nb, coords, add = TRUE)
```

```r
nb_l <- list(d1 = Sy11_nb, d2 = Sy12_nb, d3 = Sy13_nb)
sapply(nb_l, function(x) is.symmetric.nb(x, verbose = FALSE, force = TRUE))
sapply(nb_l, function(x) n.comp.nb(x)$nc)
```

## Spatial Weights

* Spatial weights can be seen as a list of weights indexed by a list of neighbours, where the weight of the link between *i* and *j* is the *kth* element of the *ith* weights list component, and *k* tells us which of the ith neighbour list component values is equal to *j*.


### Spatial Weights Styles

* If we know little about the assumed spatial process, we try to avoid moving far from the binary representation of a weight of unity for neighbours, and zero otherwise

* **nb2listw** function takes a neighbours list object and converts it into a weights object **listw**

* The default conversion style is W , where the weights for each areal entity are standardised to sum to unity; this is also often called row standardization


```r
Sy0_lw_W <- nb2listw(Sy0_nb)
Sy0_lw_W
```

```r
names(Sy0_lw_W)
names(attributes(Sy0_lw_W))
```

* For style="W" , the weights vary between unity divided by the largest and smallest numbers of neighbours, and the sums of weights for each areal entity are unity.

```r
1/rev(range(card(Sy0_lw_W$neighbours)))
summary(unlist(Sy0_lw_W$weights))
summary(sapply(Sy0_lw_W$weights, sum))
```

* Other styles is the **style="B"**  binary  retains a weight of unity for each neighbour relationship, but in this case, the sums of weights for areas differ according to the numbers of neighbour areas have

```r
Sy0_lw_B <- nb2listw(Sy0_nb, style = "B")
summary(unlist(Sy0_lw_B$weights))
summary(sapply(Sy0_lw_B$weights, sum))
```

#### General Spatial Weights

* The glist argument can be used to pass a list of vectors of general weights corresponding to the neighbour relationships to nb2listw.
* Imagine that wthe strengh of neighbour relationships decreases with distance, than we can set the weights to be proportional to the inverse distance between points representing the areas using **nbdist**
* Using lapply to invert the distances, we can obtain a different structure of spatial weights from those above.
* the default value of the argument is zero.policy=FALSE , leading to an error when given an nb argument with areas with no neighbours. Setting the argument to TRUE permits the creation of the spatial weights object, with zero weights


```r
# Get the list of distances between locations
dsts <- nbdists(Sy0_nb, coordinates(Syracuse))
# Inverte it to get the weights
idw <- lapply(dsts, function(x) 1/(x/1000))
              
# Covert it to weight object with 'B' style              
Sy0_lw_idwB <- nb2listw(Sy0_nb, glist = idw, style = "B")
              
summary(unlist(Sy0_lw_idwB$weights))
summary(sapply(Sy0_lw_idwB$weights, sum))
```

### Using Weights to Simulate Spatial Autocorrelation



```r
set.seed(987654)

# get a lengh of the same size of the Syracuse data
n <- length(Sy0_nb)

# Sample values from a normal distribution
uncorr_x <- rnorm(n)

#constructing matrix to simulate spatial autocorrelation within the invIrW function
# rho is spatial autocorrelation coefficient
rho <- 0.5

autocorr_x <- invIrW(Sy0_lw_W, rho) %*% uncorr_x
```

```r
oopar <- par(mfrow=c(1,2), mar=c(4,4,3,2)+0.1)

# Uncorrelated random variable
plot(uncorr_x, lag(Sy0_lw_W, uncorr_x), xlab="", cex.lab=0.8,
 ylab="spatial lag", main="Uncorrelated random variable", cex.main=0.8)
lines(lowess(uncorr_x, lag(Sy0_lw_W, uncorr_x)), lty=2, lwd=2)

# Uncorrelated random variable
plot(autocorr_x, lag(Sy0_lw_W, autocorr_x),
 xlab="", ylab="",
 main="Autocorrelated random variable", cex.main=0.8, cex.lab=0.8)
lines(lowess(autocorr_x, lag(Sy0_lw_W, autocorr_x)), lty=2, lwd=2)


par(oopar)
```

## Spatial Autocorrelation: Tests


* We have establish neighbours and assigned them spatial weights, we can now test for the presence of spatial autocorrelation.
* Tests assume that the mean model of the data had removed all systematic spatial patterning such as environmental drivers such as temperature, precipitation, or elevation.
* A better approach would be autocorrelation through modelling

#### Morgan's *I*

* calculated as a ratio of the product of the variable of interest and its spatial lag, with the cross-product of the variable of interest, and adjusted for the spatial weights used

<p>&nbsp;</p>

\begin{split}
I & = 
\frac{n}{\sum_{i=1}^{n} \sum_{j=1}^{n} w_{ij}}
\frac{\sum_{i=1}^{n} \sum_{j=1}^{n} w_{ij}(y_i - \bar y)(y_j - \bar y)}{\sum_{i=1}^{n} (y_i - \bar y)^2}\\
\end{split}
<p>&nbsp;</p>

```r
# Uncorrelated data
moran_u <- moran.test(uncorr_x, listw = Sy0_lw_W)
moran_u
```

```r
# Autocorrelated data
moran_a <- moran.test(autocorr_x, listw = Sy0_lw_W)
moran_a
```

```r
moran_a1 <- moran.test(autocorr_x, listw = nb2listw(Sy9_nb, style = "W"))
moran_a1
```

### Global Tests

* Morans I  **moran.test()**  is perhaps the most common global test
* Other commonly used tests include Gearys C ( **geary.test()** ), the global Getis-Ord G( **globalG.test()**), cross product Mantel test (**sp.mantel.mc()**)
* There are also join count tests for categorical variables **joincount.test()**


* The values of these statistics may be of some interest in themselves, but are not directly interpretable.

The logic of the test is the following:
1. standardize the observed value by subtracting the analytical expected value
2. divide the difference by square root of the analytical variance for the spatial weights
3. Compare with the Normal distribution to find the probability value of the observated statistic under the null hypothesis of no spatial dependence for the chosen spatial weights

```r
#autocorrelation using the default  spatial weights style of row standardisation, and using the analytical randomization assumption in computing the variance of the statistic
moran.test(NY8$Cases, listw = nb2listw(NY_nb))
```

* Changing the style of the spatial weights to make all weights equal and summing to the number of observations, we see that the resulting probability value is reduced about 20 times
* row-standardisation favours observations with few neighbours, and that styles B , C , and U weight up observations with many neighbours

```r
lw_B <- nb2listw(NY_nb, style = "B")
moran.test(NY8$Cases, listw = lw_B)
```

* By default, moran.test uses the randomisation assumption, which differs from the simpler normality assumption by introducing a correction term based on the kurtosis of the variable of interest

```r
moran.test(NY8$Cases, listw = lw_B, randomisation = FALSE)
```

* under normality is in fact the same test as the Moran test for regression residuals for the model, including only the intercept

```r
lm.morantest(lm(Cases ~ 1, NY8), listw = lw_B)
```

```r
# Saddlepoint approximation rather than the analytical normal assumption + exact test
lm.morantest.sad(lm(Cases ~ 1, NY8), listw = lw_B)

lm.morantest.exact(lm(Cases ~ 1, NY8), listw = lw_B)
```

```r
# Using a Monte Carlo permutation test
set.seed(1234)
bperm <- moran.mc(NY8$Cases, listw = lw_B, nsim = 999)
bperm
```

## Local Tests


* we can break global measures down into their components, and by extension, construct localised tests intended to detect clusters  observations with very similar neighbours and hotspots  observations with very different neighbours

* Global Morans I estimates linear relationship between variables, thus we canapply standard techniques for detecting observations with unusually strong influence on the slope.

* Local Morans I i values are constructed as the n components summed to reach global Morans I:
\
<p>&nbsp;</p>

\begin{split}
I_i & = 
\frac{(y_i - \bar y) \sum_{j=1}^{n} w_{ij}(y_j - \bar y)}{\frac{ \sum_{i=1}^{n} (y_i - \bar y)^2}{n}}\\
\end{split}
<p>&nbsp;</p>

```r
moran.plot(NY8$Cases, listw = nb2listw(NY_nb, style = "C"))
```

#### Normality, and randomisation analytically, and using Saddlepoint approximations and exact method

* Saddlepoint approximations and exact method can be important in cases where the number of neighbours of each observation is very small (nomality assuptions).


```r
# Morgan Local's test
lm1 <- localmoran(NY8$Cases, listw = nb2listw(NY_nb, style = "C"))
# Saddlepoint approximations
lm2 <- as.data.frame(localmoran.sad(lm(Cases ~ 1, NY8), nb = NY_nb, style = "C"))
# Exact
lm3 <- as.data.frame(localmoran.exact(lm(Cases ~ 1, NY8), nb = NY_nb, style = "C"))
```

```r
# Constant risk hypothesis/Poison assumption
r <- sum(NY8$Cases)/sum(NY8$POP8)
rni <- r*NY8$POP8
lw <- nb2listw(NY_nb, style="C")
sdCR <- (NY8$Cases - rni)/sqrt(rni)
wsdCR <- lag(lw, sdCR)
I_CR <- sdCR * wsdCR
```

```r
library(RColorBrewer)
gry <- c(rev(brewer.pal(8, "Reds")[1:6]), brewer.pal(6, "Blues"))
NY8$Standard <- lm1[,1]
NY8$"Constant_risk" <- I_CR
#nms <- match(c("Standard", "Constant_risk"), names(NY8))
spplot(NY8, c("Standard", "Constant_risk"), at=c(-2.5,-1.4,-0.6,-0.2,0,0.2,0.6,4,7), col.regions=colorRampPalette(gry)(8))
```

* We can uso a simple Monte Carlo test of the constant risk hypothesis local Moran'I

1. Simulate as the global case but retaining all local results
2. Extract the rank of the boserved constant risk local Moran's I
3. Calculate probability for the number of simularions made

```r
set.seed(1234)
nsim <- 999
N <- length(rni)
sims <- matrix(0, ncol=nsim, nrow=N)
for (i in 1:nsim) {
  y <- rpois(N, lambda=rni)
  sdCRi <- (y - rni)/sqrt(rni)
  wsdCRi <- lag(lw, sdCRi)
  sims[,i] <- sdCRi * wsdCRi 
}
xrank <- apply(cbind(I_CR, sims), 1, function(x) rank(x)[1])
diff <- nsim - xrank
diff <- ifelse(diff > 0, diff, 0)
pval <- punif((diff + 1)/(nsim + 1))
```

```r
NY8$Normal <- lm2[,3]
NY8$Randomisation <- lm1[,5]
NY8$Saddlepoint <- lm2[,5]
NY8$Exact <- lm3[,5]
NY8$Constant_risk <- pval
gry <- c(rev(brewer.pal(6, "Reds")), brewer.pal(6, "Blues"))
spplot(NY8, c("Normal", "Randomisation", "Saddlepoint", "Exact", "Constant_risk"), at=c(0,0.01,0.05,0.1,0.9,0.95,0.99,1), col.regions=colorRampPalette(gry)(7))
```
