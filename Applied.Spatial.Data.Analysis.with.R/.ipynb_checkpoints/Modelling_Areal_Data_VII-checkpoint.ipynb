# Modelling Areal Data


## Introduction

* The lack of independence between observations in spatial data – spatial autocorrelation – is common, and we can test it (e.g., Morgan's I)

* Often observations on relevant covariates are not available, and that the detection of spatial autocorrelation in data or model residuals is sometimes the only way left to model the remaining variation

Problems when trying to fit models in the presence of spatial
autocorrelation:
   1. spatial autocorrelation
   2. bad use of the spatial structure
       . Heteroskedasticity
       . Missing covatiate
       . Functional form of included covariates
       . Reconsidering the distribution of the response variable
       
### Spatial Statistics Approaches

* Spatial dependence can be modelled in different ways and spatial dependency (i.e.,correlation between neighbouring areas) is almost always an issue


From a statistical point of view, it is possible to account for correlated observations by:

<p>&nbsp;</p>

\begin{split}
Y & = μ + e
\end{split}
<p>&nbsp;</p>

where μ is the vector of area means, which can be modelled in different ways and e is the vector of random errors, which we assume is normally distributed with zero mean and generic variance V
* Mean is often thought to be dependent on a linear term on some covariates X, thus the mean the be substituted:

<p>&nbsp;</p>

\begin{split}
Y & = X^Tβ + e
\end{split}
<p>&nbsp;</p>


* For the case of non-Normal variables, we could transform the original data to achieve the desired Normality

* Two model will be explore to deal with correlation strutures: SAR (Simultaneous Autoregressive) and CAR (Conditionally Autoregressive) models.

The variation can be dealt with in two ways: (1) by handlying heterogeneity in different tract populations, or (2) to take a log transformation of the rate:

<p>&nbsp;</p>

\begin{split}
Z_i & = 
log \frac{1000(Y_i + 1)}{n_i}
\end{split}
<p>&nbsp;</p>

As covariates, we use:
* the inverse distanec to the closest TCE (PESPOSURE)
* The proportion of people aged 65 or higher (PCTAGE65P)
* The proportion of people who own their own home( (PCTOWNHOME)

### Simultaneous Autoregressive Models (SAR)

* SAR specification uses a regression on the values from the other areas to account for the spatial dependence

If we express the error as
<p>&nbsp;</p>

\begin{split}
e & = B(Y- X^Tβ) + ε
\end{split}
<p>&nbsp;</p>

* where B is a matrix that contains the dependence parameters $b_{ij}$ and I is the identity matrix
Then, the model can also be expressed as:


<p>&nbsp;</p>

\begin{split}
Y & = X^Tβ + B(Y- X^Tβ) + ε
\end{split}
<p>&nbsp;</p>

Hence, calculated as 

\begin{split}
(I - B)(Y- X^Tβ) =  ε
\end{split}
<p>&nbsp;</p>


* These models can be estimated efficiently by maximum likelihood using the **spautolm** command

```r
nysar <- spautolm(Z ~ PEXPOSURE + PCTAGE65P + PCTOWNHOME, data = NY8, listw = NYlistw)
summary(nysar)
```

* According to the results obtained it seems that there is significant spatial correlation in the residuals because the estimated value of λ is 0.0405 and the p-value of the likelihood ratio test is 0.0220
* The proximity to a TCE seems not to be significant
* Weighted version of these models can be fitted so that tracts are weighted proportionally to the inverse of their population size

```r
nylmw <- lm(Z ~ PEXPOSURE + PCTAGE65P + PCTOWNHOME, data = NY8, weights = POP8)
summary(nylmw)
```

```r
NY8$lmwresid <- residuals(nylmw)
lm.palette <- colorRampPalette(c("white","orange", "red"), space = "rgb")
spplot(NY8, zcol="lmresid", col.regions=lm.palette(20), main="Residual from the weighted linear model")
```

```r
lm.morantest(nylmw, NYlistw)
```

* The Moran tests for regression residuals can also be used with a weighted linear model object.
* We can check this for the SAR model too, since spautolm also takes a weights argument

```r
library("spatialreg")
nysarw <- spautolm(Z ~ PEXPOSURE + PCTAGE65P + PCTOWNHOME, data = NY8, listw = NYlistw, weights = POP8)
summary(nysarw)
```

### Conditional Autoregressive Models (CAM)

* It relies on the conditional distribution of the spatial error terms.

* Instead of the whole e −i vector, only the neighbours of area i, defined in a chosen way, are used.

<p>&nbsp;</p>

\begin{split}
e_i | e_j~i \sim N(\sum_{j~i} \frac{c_{ij}e_j}{\sum_{j\sim i}c_{ij}}, \frac{σ_{e_i}^2}{\sum_{j\sim i}c_{ij}}) 
\end{split}
<p>&nbsp;</p>

```r
nycar <- spautolm(Z ~ PEXPOSURE + PCTAGE65P + PCTOWNHOME, data = NY8, family = "CAR", listw = NYlistw)
summary(nycar)
```

* The likelihood ratio test indicates that there is significant spatial autocorrelation and the estimated value of λ is 0.0841.
* Considering a weighted regression, using the population size as weights, for the same model to account for the heterogeneous distribution of the population completely removes the spatial autocorrelation in the data

```r
nycarw <- spautolm(Z ~ PEXPOSURE + PCTAGE65P + PCTOWNHOME, data = NY8, family = "CAR", listw = NYlistw, weights = POP8)
summary(nycarw)
```

### Fitting Spatial Regression Models



* The **spautolm** function fits spatial regression models by maximum likelihood, by first finding the value of the spatial autoregressive coefficient, which maximizes the log likelihood function for the model family chosen, and then fitting the other coefficients by generalized least squares at that point





```r
# 
nysarwM <- spautolm(Z ~ PEXPOSURE + PCTAGE65P + PCTOWNHOME, data = NY8, family = "SAR", listw = NYlistw, weights = POP8, method = "Matrix")
summary(nysarwM)
```

```r
# range of the L-Likelihood values for a range of values of λ
1/range(eigenw(NYlistw))

# for unweighted
nysar_ll <- spautolm(Z ~ PEXPOSURE + PCTAGE65P + PCTOWNHOME, data = NY8, family = "SAR", listw = NYlistw, llprof = 100)

# and weighted model
nysarw_ll <- spautolm(Z ~ PEXPOSURE + PCTAGE65P + PCTOWNHOME, data = NY8, family = "SAR", listw = NYlistw, weights = POP8, llprof = 100)

```

```r
ylim <- range(c(nysarw_ll$llprof$ll, nysar_ll$llprof$ll), na.rm=TRUE)
plot(nysarw_ll$llprof$lambda, nysarw_ll$llprof$ll, type="l", xlab=expression(lambda), ylab="log likelihood", ylim=ylim, lwd=2)
abline(v=nysarw_ll$lambda)
abline(h=nysarw_ll$LL)
lines(nysar_ll$llprof$lambda, nysar_ll$llprof$ll, lty=2, lwd=2)
abline(v=nysar_ll$lambda, lty=2)
abline(h=nysar_ll$LL, lty=2)
legend("bottom", legend=c("weighted SAR", "SAR"), lty=c(1,2), lwd=2, bty="n")
```

```r
# family="SMA"

nysmaw <- spautolm(Z ~ PEXPOSURE + PCTAGE65P + PCTOWNHOME, data = NY8, family = "SMA", listw = NYlistw, weights = POP8)
summary(nysmaw)
```

## Mixed-Effects Models

* The errors e i which appear in the previous models are used to account for between-area variation, following a specified correlation structure (AKA **random effects**). 
* The covariates are on the other hand are **fixed effects**.


<p>&nbsp;</p>

\begin{split}
Y & = Xβ + Ze + ε
\end{split}
<p>&nbsp;</p>

* Vector *e* represents the random effects, whilst Z is used to account for their structure.
* Z also depends on λ, which is another parameter to be estimated
* Maximum Likelihood or Restricted Maximum Likelihood and can be found in **nlme** and **lme4** package to fit that kind of models
* those models allow the specification of different types of covariance matrices of the random effects, including spatial structure

Steps:
1. Specify the correlation structure between the areas, i.e., using Gaussian vaiogram based on the Euclidean distances between centroids of the regions

```r
library(nlme)


NY8$x <- coordinates(NY8)[, 1]/1000
NY8$y <- coordinates(NY8)[, 2]/1000

# specified the correlation structure
sp1 <- corSpatial(1, form = ~x + y, type = "gaussian")
scor <- Initialize(sp1, as(NY8, "data.frame")[, c("x", "y")], nugget = FALSE)

# fixed part of the model is as in the previous
# random part of the model we need to include a random effect per area


spmodel <- lme(Z ~ PEXPOSURE + PCTAGE65P + PCTOWNHOME, random = ~1 | AREAKEY, data = as(NY8, "data.frame"), correlation = scor, method = "ML")
summary(spmodel)
```
